```python
import numpy as np
#two different conditions we are looking at - status of health and status of vaccination
allhumans=[]
healthy_a=['sick','not sick']
vax_a=['vaccinated','not vaccinated']
#define a class of people that have a name, position in terms of x and y as, health and vaccination status
class people():
    def __init__(self,name,x,y,vax,health):
        self.name=name
        self.x=x
        self.y=y
        self.health=health
        self.vax=vax
        self.mood=''
        self.info=self.x,self.y,self.health,self.vax
#given a room of size 50x50 with 100 people, I randomize their positions and statuses.
for i in range(100):
    humans=people(name=i,x=np.random.randint(50),y=np.random.randint(50),health=np.random.choice(healthy_a),vax=np.random.choice(vax_a))
    allhumans.append(humans)                                                                                                                                                                                
sick=0
sick_notvax=0
sick_vax=0
#defining a function called move, that will move a person with a given speed through a randomly assigned angle 
#and all negative positions from the varying of angles are called 0. This information is then assigned to the object of 
#humans.
def move(self,speed,angle):
    xpos=speed*np.cos(angle)
    if xpos <0:
        xpos=0
    self.x=xpos
    ypos=speed*np.sin(angle)
    if ypos <0:
        ypos=0
    self.y=ypos
    self.info=self.x,self.y,self.health,self.vax
a_sick=[]
a_nv=[]
a_v=[]
#this for loop runs through 200 times, moving all the people everytime and at the end of it generating the number of 
#people who became infected and which of these had vaccines and which didn't. So we begin by moving all the people
#with a randomly assigned speed and anled and the distance between any two people are calculated. If the distance
#is below 10 units but greated than 0 units(to avoid looking at the same person), and one of the person is sick and the
#other is not sick, a not vaccinated person would have a 60% chance of also becoming sick and a vaccinated person would
#have a 5% chance of getting sick. 
for k in range(200):
    for i in range(100):
        speed=np.random.random()*10
        angle=np.random.random()*2*np.pi
        move(allhumans[i],speed,angle)
        for j in range(100):
            d= np.sqrt((allhumans[i].x-allhumans[j].x)**2+(allhumans[i].y-allhumans[j].y)**2)
            if d<10:
                if d>0:
                    if allhumans[i].health=='sick':
                        if allhumans[j].vax=='not vaccinated':
                            prob =np.random.random()*1
                            if prob<0.6:
                                allhumans[j].health='sick'
                        else:
                            prob =np.random.random()*1
                            if prob<0.05:
                                allhumans[j].health=='sick'
       
#this saves the number of people who are sick and those of which are also vaccinated or not vaccinated and prints out
#the number
sicka=[]
sick=0
sick_vax=0
sick_vaxa=[]
sick_notvax=0
sick_notvaxa=[]
for i in range(100):
    if allhumans[i].health=='sick':
        sick=sick+1
        sicka.append(allhumans[i].name)
        if allhumans[i].vax=='vaccinated':
            sick_vax=sick_vax+1
            sick_vaxa.append(allhumans[i].name)
        else:
            sick_notvax=sick_notvax+1
            sick_notvaxa.append(allhumans[i].name)
print('the total number of sick people in the room',sick)
print('the number of sick people who were not vaccinated',sick_notvax)
print('the number of sick people who were vaccinated',sick_vax)
```

```python
import matplotlib.pyplot as plt
x1=np.linspace(0,1,len(sick_notvaxa))
x2=np.linspace(0,1,len(sick_vaxa))
x3=np.linspace(0,1,len(sicka))
plt.plot(x1,label='sick without vaccination')
plt.plot(x2,label='sick with vaccination')
plt.plot(x3,label='total sick')
plt.ylabel('np.linspace ranging')
plt.xlabel('number of people')
plt.title('comparing the total number of sick people')
plt.legend()
plt.show()
```
